```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
#Set seed for reproducibility
np.random.seed(42)

#Generate 300 lead records
n_records = 300

start_date = datetime(2024,1,1)
dates = [start_date + timedelta(days=int(x)) for x in np.random.randint(0,180,n_records)] #max=180: 6-month time window

data = {
    'Lead_ID': [f'LEAD-{i:04d}' for i in range(1, n_records + 1)],
    'Sales_Rep': np.random.choice(['Ahmad Taro', 'Siti Fatimah', 'Raj Kumar', 'Chen Wei'], n_records),
    'Project': np.random.choice(['Sentosa', 'Empire Subang', 'Southbank'], n_records),
    'Location': np.random.choice(['KL', 'Selangor', 'Penang'], n_records),
    'Lead_Date': dates,
    'Unit_Type': np.random.choice(['Residential', 'Commercial'], n_records),
    'Lead_Value': np.random.uniform(300000,1500000, n_records),
    'Status': np.random.choice(['Lead','Qualified','Proposal','Sold','Lost'], n_records)
}

df = pd.DataFrame(data)
```
