```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
#Set seed for reproducibility
np.random.seed(42)

#Generate 300 lead records
n_records = 300

start_date = datetime(2024,1,1)
dates = [start_date + timedelta(days=int(x)) for x in np.random.randint(0,180,n_records)] #max=180: 6-month time window

data = {
    'Lead_ID': [f'LEAD-{i:04d}' for i in range(1, n_records + 1)],
    'Sales_Rep': np.random.choice(['Ahmad Taro', 'Siti Fatimah', 'Raj Kumar', 'Chen Wei'], n_records),
    'Project': np.random.choice(['Sentosa', 'Empire Subang', 'Southbank'], n_records),
    'Location': np.random.choice(['KL', 'Selangor', 'Penang'], n_records),
    'Lead_Date': dates,
    'Unit_Type': np.random.choice(['Residential', 'Commercial'], n_records),
    'Lead_Value': np.random.uniform(300000,1500000, n_records),
    'Status': np.random.choice(['Lead','Qualified','Proposal','Sold','Lost'], n_records)
}

df = pd.DataFrame(data)
```

Above script doesnt have transition of stages (lead>qualified..)

```python
n_leads = 100
records = []

for lead_id in range(1,n_leads+1):
    sales_rep = np.random.choice(['Ahmad Taro', 'Siti Fatimah', 'Raj Kumar', 'Chen Wei'])
    project = np.random.choice(['Sentosa', 'Empire Subang', 'Southbank'])
    location = np.random.choice(['KL', 'Selangor', 'Penang'])
    start_date = datetime(2024,1,1) + timedelta(days=np.random.randint(0,180))
    unit_value = np.random.uniform(300000,1500000)

    #Determine final outcome
    outcome = np.random.choice(['Sold', 'Lost', 'Stalled'], p=[0.20,0.25,0.55])

    #Create progression based on outcome
    if outcome == 'Sold':
        stages = [
            ('Lead', start_date),
            ('Qualified', start_date + timedelta(days=np.random.randint(3,10))),
            ('Proposal', start_date + timedelta(days=np.random.randint(15,25))),
            ('Sold', start_date + timedelta(days=np.random.randint(30,60)))
        ]
    
    elif outcome == 'Lost':
        stages = [
            ('Lead', start_date),
            ('Qualified', start_date + timedelta(days=np.random.randint(3,10))),
            ('Lost', start_date + timedelta(days=np.random.randint(15,30))),            
        ]

    else:
        stages = [
            ('Lead', start_date),
            ('Qualified', start_date + timedelta(days=np.random.randint(3,10))),
            ('Proposal', start_date + timedelta(days=np.random.randint(15,25))),
        ]
    
    for stage, date in stages:
        records.append({
            'Lead_ID': f'LEAD-{lead_id:04d}',
            'Sales_Rep': sales_rep,
            'Project': project,
            'Location': location,
            'Stage': stage,
            'Date': date,
            'Unit_Value': unit_value
        })

df2 = pd.DataFrame(records)
df2.to_clipboard()
```
